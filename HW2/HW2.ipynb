```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import pandas as pd
data=pd.read_csv('games.data.csv')
data.head()
```

```python
#
data=pd.read_csv('games.data.csv')

plat = data['Platform'].value_counts()
plat.plot.barh(figsize=(15,10), xlabel='Platform', title='Platform Ranking (per game)');
```

```python
#
import matplotlib.font_manager as fm

plat = data['Year_of_Release'].value_counts()
plat.plot.bar(figsize=(15,10), xlabel='Year', title='Year_of_Release Ranking (per year)');
```

```python
#
most_common_G = data['Genre'].value_counts().idxmax()
value_counts_G= data['Genre'].value_counts()
print("")
print(value_counts_G)
print("Genre", most_common_G)
```
