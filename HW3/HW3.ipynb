```python
#標題一樣的問題解決 嘗試給標題作頁數標籤
urlfirst = 'https://www.ptt.cc/bbs/Nogizaka46/index'
pages = range(1, 10)
urlsec = '.html'

titles = []
pagelist=[]

for page in pages:
    link = urlfirst + str(page) + urlsec
    response = requests.get(link)
    response.encoding = 'utf-8'
    soup = BeautifulSoup(response.text, "html.parser")
    title_tags = soup.find_all(class_='title')
    for title in title_tags:
        titles.append(title.text)
        
for title in titles:
    print(title)
```

```python
#頁數標籤完成! お疲れ！
urlfirst = 'https://www.ptt.cc/bbs/Nogizaka46/index'
pages = range(1, 10)
urlsec = '.html'

titles = []

for page in pages:
    link = urlfirst + str(page) + urlsec
    response = requests.get(link)
    response.encoding = 'utf-8'
    soup = BeautifulSoup(response.text, "html.parser")
    title_tags = soup.find_all(class_='title')
    for title in title_tags:
        titles.append(title.text)
x=20
for title in titles:
    if x%20==0:
        print("page:",int(x/20)) #每頁20條文
    print(int(x/20),"-",x%20+1,title)
    x+=1
    
```
